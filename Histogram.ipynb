```python
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np 
import glob
```

```python
W_abnormal = []
H_abnormal = []
W_normal = []
H_normal = []
```

```python
for filename in glob.glob('PartA_DFU_Dataset/Abnormal/*.jpg'): 
    im = Image.open(filename)
    W_abnormal.append(im.width)
    H_abnormal.append(im.height)
```

```python
for filename in glob.glob('PartA_DFU_Dataset/Normal/*.jpg'): 
    im = Image.open(filename)
    W_normal.append(im.width)
    H_normal.append(im.height)
```

```python
abnormal_size = np.multiply(W_abnormal, H_abnormal)
normal_size = np.multiply(W_normal,H_normal)
```
