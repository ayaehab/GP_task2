```python
    
model1=Sequential([
    # input
#     BatchNormalization(),
    ReLU(),
    Conv2D(filters=64, kernel_size=(3,3), strides=1, activation='relu',padding='same'),
    Dropout(0.25),
    # first layer
    Conv2D(filters=64, kernel_size=(7,7), strides=2, activation='relu',padding='same'),
    MaxPool2D(pool_size=3, strides=2,padding='same'),
    
    # adding dense block 1
   
    Conv2D(filters=64, kernel_size=(1,1), activation='relu',padding='same'),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same'),
    
    Conv2D(filters=64, kernel_size=(1,1), activation='relu',padding='same'),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same'),
    
    Conv2D(filters=64, kernel_size=(1,1), activation='relu',padding='same'),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same'),
    
    Conv2D(filters=64, kernel_size=(1,1), activation='relu',padding='same'),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same'),
    
    Conv2D(filters=64, kernel_size=(1,1), activation='relu',padding='same'),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same'),
    
    Conv2D(filters=64, kernel_size=(1,1), activation='relu',padding='same'),
    Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same')    
])
```

```python
# model1.build([None,224, 224, 3])
# model1.summary()
```

```python
trans1=model1
# add transition layer 1
trans1.add(ReLU())
trans1.add(Conv2D(filters=64, kernel_size=(1,1), activation='relu'))
trans1.add(Dropout(0.1))
trans1.add(AveragePooling2D(pool_size=(2, 2), strides=2,padding='same'))
# trans1.build([None,224, 224, 3])
# trans1.summary()
```

```python
#add dense block 2
model2=trans1
model2.add(Conv2D(filters=64, kernel_size=(1,1), activation='relu'))
model2.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same'))

for i in np.arange(11):
    model2.add(Conv2D(filters=64, kernel_size=(1,1), activation='relu',padding='same'))
    model2.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same'))
    
model2.build([None,224, 224, 3])
model2.summary()
```

```python
trans2=model2
# add transition layer 2
trans2.add(ReLU())
trans2.add(Conv2D(filters=64, kernel_size=(1,1), activation='relu'))
trans2.add(Dropout(0.1))
trans2.add(AveragePooling2D(pool_size=(2, 2), strides=2,padding='same'))
trans2.build([None,224, 224, 3])
trans2.summary()
```

```python
#add dense block 3
model3=trans2

model3.add(Conv2D(filters=64, kernel_size=(1,1), activation='relu'))
model3.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same'))

for i in np.arange(23):
    model3.add(Conv2D(filters=64, kernel_size=(1,1), activation='relu',padding='same'))
    model3.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same'))
    
model3.build([None,224, 224, 3])
model3.summary()
```

```python
#add transition layer 3
trans3=model3
trans3.add(ReLU())
trans3.add(Conv2D(filters=64, kernel_size=(1,1), activation='relu'))
trans3.add(Dropout(0.1))
trans3.add(AveragePooling2D(pool_size=(2, 2), strides=2,padding='same'))
trans3.build([None,224, 224, 3])
trans3.summary()
```

```python
#add dense block 4
model4= trans3


model4.add(Conv2D(filters=64, kernel_size=(1,1), activation='relu'))
model4.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same'))

for i in np.arange(15):
    model4.add(Conv2D(filters=64, kernel_size=(1,1), activation='relu',padding='same'))
    model4.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu',padding='same'))
    
model4.build([None,224, 224, 3])
model4.summary()
```

```python
model4.add(GlobalAveragePooling2D(data_format='channels_last'))
    # full connection
model4.add(Dense(units=1000, activation='softmax'))
model4.build([None,224, 224, 3])
model4.summary()
```

```python
model4.compile(optimizer='adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model4.fit(x=training_set, validation_data=testing_set, epochs=30)
```
